```python
import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
mf = [(2.75, 11.7), (3.34, 19.7), (8.55, 37.6), (20.82, 75.2), (26.17, 143.9), (28.84, 182.8), (32.50, 222.8) ]
inavit = [(3.19, 14.4), (4.64, 23.4), (10.41, 41.0), (26.86, 83.3), (32.24, 157.2), (34.58, 194.1), (39.89, 238.7)]

fig = plt.figure(figsize=(7.02,3.54), dpi=600)
ax = fig.add_subplot(111)
layers = [1,2,4,8,12,16,20]

mf_x = [yt for xt,yt in mf]
mf_y = [xt for xt,yt in mf]
in_x = [yt for xt,yt in inavit]
in_y = [xt for xt,yt in inavit]

plt.plot(mf_x, mf_y, "b--", label="MotionFormer")
plt.plot(in_x, in_y, "r--", label="InAViT")
plt.legend(loc="upper left")
plt.xlabel('Number of Parameters', fontsize=15)
plt.ylabel('Action Mean Recall @5', fontsize=15)
for yt, xt in mf:
    ax.scatter(xt, yt, color='blue', marker='o')
    
for i in range(len(mf)):
    plt.annotate(layers[i], (mf[i][1],mf[i][0]), xytext=(5, 5), textcoords="offset pixels")

for yt, xt in inavit:
    ax.scatter(xt, yt, color='red', marker='*')
for i in range(len(inavit)):
    plt.annotate(layers[i], (inavit[i][1],inavit[i][0]), xytext=(5, 5), textcoords="offset pixels")

plt.show()
```
